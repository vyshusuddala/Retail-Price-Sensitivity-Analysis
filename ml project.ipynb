```python
# =========================================================
# Retail Price Sensitivity Analysis
# Technologies: MS Excel, Python, statsmodels, matplotlib
# Environment: Google Colab
# =========================================================

import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import os

# ---------------------------------------------------------
# Step 1: CREATE Excel File (MS Excel usage - Input)
# ---------------------------------------------------------
# This avoids FileNotFoundError in Google Colab

data = {
    "Price": [50, 55, 60, 65, 70, 75, 80],
    "Quantity_Sold": [500, 470, 430, 380, 320, 260, 200]
}

input_df = pd.DataFrame(data)
input_df.to_excel("retail_data.xlsx", index=False)

print("✅ retail_data.xlsx created successfully")

# ---------------------------------------------------------
# Step 2: LOAD Data from Excel (MS Excel → Python)
# ---------------------------------------------------------
df = pd.read_excel("retail_data.xlsx")

print("\nInitial Data:")
print(df)

# ---------------------------------------------------------
# Step 3: Data Cleaning & Preparation (Python)
# ---------------------------------------------------------
df = df.dropna()
df = df.drop_duplicates()

df["Price"] = df["Price"].astype(float)
df["Quantity_Sold"] = df["Quantity_Sold"].astype(float)

# ---------------------------------------------------------
# Step 4: Exploratory Data Analysis (matplotlib)
# ---------------------------------------------------------
plt.figure()
plt.scatter(df["Price"], df["Quantity_Sold"])
plt.xlabel("Price")
plt.ylabel("Quantity Sold")
plt.title("Price vs Quantity Sold")
plt.show()

# ---------------------------------------------------------
# Step 5: Price Sensitivity Regression (statsmodels)
# --------------------
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset with different experience levels
data = {
    "Role": [
        "Data Analyst", "Data Analyst", "Data Analyst",
        "Data Scientist", "Data Scientist", "Data Scientist",
        "Software Engineer", "Software Engineer", "Software Engineer", "Software Engineer",
        "Business Analyst", "Business Analyst"
    ],
    "Experience": [0, 2, 4, 0, 3, 6, 0, 1, 3, 5, 1, 4],
    "Salary": [3, 6, 10, 6, 14, 24, 4, 6, 11, 18, 5, 9]
}

df = pd.DataFrame(data)

print(df)

# ---------------------------------
# Bar Chart: Average Salary by Role
# ---------------------------------
avg_salary = df.groupby("Role")["Salary"].mean()

plt.figure()
plt.bar(avg_salary.index, avg_salary.values)
plt.xlabel("Job Role")
plt.ylabel("Average Salary (LPA)")
plt.title("Average Salary by Job Role")
plt.show()

# ---------------------------------
# Box Plot: Salary Distribution by Role
# ---------------------------------
plt.figure()
df.boxplot(column="Salary", by="Role")
plt.xlabel("Job Role")
plt.ylabel("Salary (LPA)")
plt.title("Salary Distribution by Job Role")
plt.suptitle("")
plt.show()

# ---------------------------------
# Scatter Plot: Experience vs Salary
# ---------------------------------
plt.figure()
plt.scatter(df["Experience"], df["Salary"])
plt.xlabel("Experience (Years)")
plt.ylabel("Salary (LPA)")
plt.title("Experience vs Salary Across Roles")
plt.show()
```
